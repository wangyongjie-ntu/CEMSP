```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
mu1 = [0, 0]
cov1 = [[4, 2], [2, 1]]
A, B = np.random.multivariate_normal(mu1, cov1, 5000).T
plt.plot(A, B, 'x')
#plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt
#mu, sigma = 0, 0.1 # mean and standard deviation
mu2 = 5
sigma2 = 2
C = np.random.normal(mu2, sigma2, 5000)
count, bins, ignored = plt.hist(C, 30, density=True)
plt.plot(bins, 1/(sigma2 * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu2)**2 / (2 * sigma2**2) ),
         linewidth=2, color='r')
plt.show()
```

```python
D = C + 10
count, bins, ignored = plt.hist(D, 30, density=True)
plt.plot(bins, 1/(sigma2 * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu2 - 10)**2 / (2 * sigma2**2)),
         linewidth=2, color='r')
plt.show()
```

```python
low = 0
high = 50
E = np.random.uniform(low, high, 5000)
count, bins, ignored = plt.hist(E, 50, density=True)
plt.plot(bins[0:-1], count,
         linewidth=2, color='r')
plt.show()
```
